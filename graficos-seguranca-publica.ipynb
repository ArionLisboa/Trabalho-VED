## Reordenação dos dados para apresentar diferentes pontos de vista

### Códigos prontos

```python
# Criando uma lista com todas as Uniões Federativas do banco de dados

ufs = []
for uf in segpub["UF"]:
    if not(uf in ufs):
        ufs.append(uf)

ufs
```

```python
# Criando uma lista com todos os anos presentes na base de dados

anos = []
for ano in segpub["Ano"]:
    if not(ano in anos):
        anos.append(ano)

anos
```

```python
# Criando uma lista com todos os crime apresentados na base de dados

crimes = {}
for crime in segpub["Tipo Crime"]:
    if not(crime in crimes.keys()):
        crimes[crime] = []

crimes
```

### Reordenação

```python
ano2021 = segpub[segpub["Ano"] == 2021]
```

```python
crimes21 = {}
i = 0
for crime in ano2021["Tipo Crime"]:
    if crime in crimes21.keys():
        crimes21[crime] += ano2021.iloc[i]["Ocorrências"]
    else:
        crimes21[crime] = ano2021.iloc[i]["Ocorrências"]
    i += 1
```
